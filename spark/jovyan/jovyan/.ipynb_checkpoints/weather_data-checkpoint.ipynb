```python
import pandas as pd
import requests
import calendar
import datetime
from datetime import datetime
import numpy as np
from credentials import weather_api_key

train_df = pd.read_csv('rail_data_eda.csv')
train_df.head()
```

```python
dates = train_df['service_start_date'].unique()
dates
```

```python
api_key = weather_api_key

# get approx lat and long for each region
regions = {
    'Northern': {'Latitude': 55.0, 'Longitude': -2.0},
    'Southern': {'Latitude': 50.5, 'Longitude': -1.5},
    'Eastern': {'Latitude': 53.0, 'Longitude': 0.0},
    'Western': {'Latitude': 53.0, 'Longitude': -4.0},
    'Central': {'Latitude': 52.0, 'Longitude': -1.5}
}

urls = []
for i in dates:
    date_obj = datetime.strptime(i, '%Y-%m-%d')
    date = int(date_obj.timestamp())
    
    urls.append(f'https://api.openweathermap.org/data/2.5/onecall/timemachine?lat=55.0&lon=-2.0&dt={date}&appid={api_key}')
    urls.append(f'https://api.openweathermap.org/data/2.5/onecall/timemachine?lat=50.5&lon=-1.5&dt={date}&appid={api_key}')
    urls.append(f'https://api.openweathermap.org/data/2.5/onecall/timemachine?lat=53.0&lon=0.0&dt={date}&appid={api_key}')
    urls.append(f'https://api.openweathermap.org/data/2.5/onecall/timemachine?lat=53.0&lon=-4.0&dt={date}&appid={api_key}')
    urls.append(f'https://api.openweathermap.org/data/2.5/onecall/timemachine?lat=52.0&lon=-1.5&dt={date}&appid={api_key}')

# call data for each url and combine
data = []
for url in urls:
    response = requests.get(url)
    data.append(response.json())

d = {
    'id' : [], 
    'name' : [], 
    'dt' : [],
    'longitude' : [], 
    'latitude' : [], 
    'temp' : [], 
    'temp': [],
    'feels_like': [],
    'temp_min': [],
    'temp_max': [],
    'pressure': [],
    'sea_level': [],
    'grnd_level': [],
    'humidity': [],
    'weather' : [],
    'weather_desc': []
    }

for i in data:
    d['id'].append(i['id'])
    d['name'].append(i['name'])
    d['dt'].append(i['dt'])
    d['longitude'].append(i['coord']['lon'])
    d['latitude'].append(i['coord']['lat'])
    d['temp'].append(i['main']['temp'])
    d['feels_like'].append(i['main']['feels_like'])
    d['temp_min'].append(i['main']['temp_min'])
    d['temp_max'].append(i['main']['temp_max'])
    d['pressure'].append(i['main']['pressure'])
    d['sea_level'].append(i['main']['sea_level'])
    d['grnd_level'].append(i['main']['grnd_level'])
    d['humidity'].append(i['main']['humidity'])
    d['weather'].append(i['weather'][0]['main'])
    d['weather_desc'].append(i['weather'][0]['description'])
```

```python
df = pd.DataFrame(d)
df.head()
```

```python
df['dt'].unique()
```

```python
def add_weather_id(region):
    coords = regions[region]
    for i, row in df.iterrows():
        if row['longitude'] == coords['Longitude'] and row['latitude'] == coords['Latitude']:
            return row['id']
```

```python
train_df['region_weather_id'] = train_df['Region'].apply(add_weather_id)
train_df.head()
```

```python
merged_df = train_df.merge(df, left_on='region_weather_id', right_on='id')
merged_df.head()
```

```python
merged_df.to_csv('weather_rail_data.csv', index=False)
```
